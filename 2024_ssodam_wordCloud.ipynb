## 문자열을 리스트로 변환

```python
def get_noun_list_from_str(msg):
    import re
    try :
        return [tuple([re.sub("'",'',y) for y in re.findall('\'.*?\'',x)]) for x in re.findall('\(.*?\)',msg)]
    except :
        return np.NaN
```

```python
df = df.fillna('')
df.contents_tokenized = df.contents_tokenized.apply(lambda x: get_noun_list_from_str(x))
df.title_tokenized = df.title_tokenized.apply(lambda x: get_noun_list_from_str(x))
df.comments_tokenized = df.comments_tokenized.apply(lambda x: get_noun_list_from_str(x))
df.date = pd.to_datetime(df.date)
```

```python
examples = open('/content/drive/MyDrive/stopWords.txt','r',encoding='utf-8')
examples_lst = examples.read().splitlines()
```

```python
stopwords = ['의','가','이','은','들','는','좀','잘','걍','과','도','를','으로','을','안','한','로',',','아','다미','부터','이랑',
             '자','에','와','하다','ㅋ','명','것','곳','거','때','그','데','못','중','대한','인데','진짜','너무','에서','근데',
             '000', '원', '데헷']


okt = Okt()
extra = '캡쳐 및 복사 금지 해당 게시물은 서강대 동문만을 위해 작성된 게시물로 외부로 유출시 서담 영구 탈퇴되며 저작권 침해 개인정보보호법 위반 등의 법적 책임을 질 수 있습니다'
extra_lst1 = okt.morphs(extra)
extra_lst2 = extra.split() #보조장치

stopwords += extra_lst1
stopwords += extra_lst2
stopwords += examples_lst
stopwords = list(set(stopwords))
```

```python
df.contents_tokenized = df.contents_tokenized.apply(lambda x: [item for item in x if len(item) == 2])
df.title_tokenized = df.title_tokenized.apply(lambda x: [item for item in x if len(item) == 2])
df.comments_tokenized = df.comments_tokenized.apply(lambda x: [item for item in x if len(item) == 2])
```

# 단어 묶기

```python
# 앞 뒤 토큰을 묶어서 합칠 수 있으면 하나의 단어로 묶어서 인식시키는 함수 (ex. 서강 + 대학교)
def concat(x):
    w = []
    for i in range(len(x)):
        if x[i][1] == 'Noun' or x[i][1] == 'Number':
            try:
                if x[i-1][1] == 'Modifier':
                    pass
                else:
                    w.append(x[i][0])
            except:
                pass
        elif x[i][1] == 'Suffix':
            try:
                w.pop()
                w.append(x[i-1][0] + x[i][0])
            except:
                pass
        elif x[i][1] == 'Modifier':
            try:
                w.append(x[i][0] + x[i+1][0])
            except:
                pass
    return w
```

```python
tqdm.pandas() #진행률 프로세스바
df['contents_w'] = df.contents_tokenized.progress_apply(lambda x: concat(x))
df['title_w'] = df.title_tokenized.progress_apply(lambda x: concat(x))
```

# StyleCloud 시각화

```python
stopwords_ = pd.read_csv('/content/drive/MyDrive/stopwords.txt', encoding='UTF-8', header=None)
stopwords_ = set(stopwords_[0].values.tolist())
stopwords__ = {'부연위원','드릴예정','충성','더리움','생각함','2외국어','구성위','차촬영','서강49','료지원','차면접','위원19년','심폐지','폐지구력','달정도' '부연위원','연구원부연','차합격자','생인턴','서강학','관심분','분요','성바이오피스','톡실행','아시분','니사례','녹화구합','애', '구합니', '기보본강의', '양도합도합니', '대해공부', '학년1', '점만점', '알도움', '공유분','차시험', '이기때문', '츠벤또', '세입문', '업종마카', '맞이사업', '맞이단', '새내기위백', '카페21', '컴공복', '담시방'}
stopwords_ = stopwords_.union(stopwords__)
stopwords_
```

```python
def style_cloud(board, icon):
    target = df
    tokenized_contents = [
        list(zip((sentence), (sentence)[1:]))
        for sentence in target.contents_w.values
    ]  # text_W 토큰화

    tokenized_title = [
        list(zip((sentence), (sentence)[1:]))
        for sentence in target.title_w.values
    ]  # title_w 토큰화해서 합치기

    tokenized_nouns = tokenized_contents + tokenized_title

    counter = Counter()
    for i in tokenized_nouns:
        counter.update(Counter(i))
    mc = counter.most_common()

    wdic = {}
    for i in mc:
        if i[0][0] != i[0][1]:
            wdic[i[0][0] + i[0][1]] = i[1]
        else:
            wdic[i[0][0]] = i[1]

    if '담시방' in wdic:
        wdic['서담시방'] = wdic['담시방']
        del wdic['담시방']

    if '워밸' in wdic:
        wdic['워라밸'] = wdic['워밸']
        del wdic['워밸']

    if '/' in board:
        board = board.split('/')
        fname = 'fig' + board[0] + '.png'
    else:
        fname = 'fig' + str(board) + '.png'

    stylecloud.gen_stylecloud(text=wdic,
                              icon_name=icon,
                              font_path='/content/drive/MyDrive/NanumSquareB.ttf',
                              palette='cartocolors.qualitative.Prism_10',
                              background_color="rgba(255, 255, 255, 0)",
                              gradient="horizontal",
                              custom_stopwords=stopwords_,
                              output_name=fname)

    return
```

```python
style_cloud("CPA", "fas fa-calculator")
```
